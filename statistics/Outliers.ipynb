```python
import numpy as np

```

```python
arr = np.array([2,4,7,10,15,20,25,205])
arr
```

#### Quartiles

```python
Q1 = np.percentile(arr,25)
Q1
```

```python
Q3 = np.percentile(arr,75)
Q3
```

```python
print(Q1)
```

```python
print(Q3)
```

#### IQR

```python
IQR = Q3 - Q1
print(IQR)
```

#### Lower and Upper fences


```python
lower_fence = Q1 - 1.5 * IQR
upper_fence = Q3 + 1.5 * IQR
```

```python
print(lower_fence)
```

```python
print(upper_fence)
```

#### Detect outliers


```python
Outliers = arr[(arr<lower_fence) | (arr > upper_fence)]
Outliers
```

#### Remove outliers


```python
clean_data = arr[(arr >= lower_fence) & (arr <= upper_fence)]
clean_data
```
