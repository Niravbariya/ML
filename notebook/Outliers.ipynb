```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
arr = np.array([2,4,7,10,15,20,25,205])
print(arr)
```

```python
s = pd.Series(arr)
s
```

```python
sns.boxplot(x=s)
```

```python
s.describe()
```

```python
s.median()
```

```python
print(s.sort_values())
```

```python
q1 = s.quantile(0.25)
q1
```

```python
q3 = s.quantile(0.75)
q3
```

```python
IQR = q3 - q1
print(IQR)
```

```python
lower_fence = q1 - 1.5 * IQR
print(lower_fence)
```

```python
upper_fence = q3 + 1.5 * IQR
print(upper_fence)
```

```python
outliers = s[(s < lower_fence) | (s > upper_fence)]

print("Outliers:")
print(outliers)
```

```python
clean_data = s[(s >= lower_fence) & (s <= upper_fence)]

print(clean_data)
```

```python
sns.boxplot(x = clean_data)
```
